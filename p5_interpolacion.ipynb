```python
def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float]:
    c2 = sum(x**2 for x in xs)
    c1 = sum(x for x in xs)
    c0 = len(xs)
    c_ind = sum(ys)
    return c2, c1, c0, c_ind


def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float]:
    c2 = sum(x**3 for x in xs)
    c1 = sum(x**2 for x in xs)
    c0 = sum(x for x in xs)
    c_ind = sum(x*y for x, y in zip(xs, ys))
    return c2, c1, c0, c_ind


def der_parcial_2(xs: list, ys: list) -> tuple[float, float, float]:
    c2 = sum(x**4 for x in xs)
    c1 = sum(x**3 for x in xs)
    c0 = sum(x**2 for x in xs)
    c_ind = sum((x**2)*y for x, y in zip(xs, ys))
    return c2, c1, c0, c_ind
```

```python
from src import ajustar_min_cuadrados

xs1 = [
    -5.0000,
    -3.8889,
    -2.7778,
    -1.6667,
    -0.5556,
    0.5556,
    1.6667,
    2.7778,
    3.8889,
    5.0000,
]
ys1 = [
    57.2441,
    33.0303,
    16.4817,
    7.0299,
    0.5498,
    0.7117,
    3.4185,
    12.1767,
    24.9167,
    44.2495,
]

x1,x2,x3 = ajustar_min_cuadrados(xs1, ys1, [der_parcial_0, der_parcial_1,der_parcial_2])
print("Valores obtenidos")
print("a: {:.4e}".format(x1))
print("b: {:.4e}".format(x2))
print("c: {:.4e}".format(x3))
```

```python
# Reemplace con la función obtenida de ajustar_min_cuadrados
import numpy as np
import matplotlib.pyplot as plt
def interpolacion(x): return x1*x**2+x2*x+x3
# --- Función de ejemplo


# Dibujar los puntos x, y
plt.scatter(xs1, ys1, alpha=0.5, label="Datos")

# Dibujar la función de aproximación
x_aprox = np.linspace(min(xs1), max(xs1), 100)
y_aprox = [interpolacion(xi) for xi in x_aprox]
plt.plot(x_aprox, y_aprox, "r-", label="Interpolación: y = ")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
def der_parcial_0_exp(xs: list, ys: list) -> tuple[float, float, float]:
    c1 = sum(x for x in xs)

    c0 = len(xs)

    c_ind = sum(math.log(y) for y in ys)

    return c1, c0, c_ind


def der_parcial_1_exp(xs: list, ys: list) -> tuple[float, float, float]:
    c1 = sum(x**2 for x in xs)

    c0 = sum(x for x in xs)

    c_ind = sum(x * math.log(y) for x, y in zip(xs, ys))

    return c1, c0, c_ind
```
